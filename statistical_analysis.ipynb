## Importing Required Modules

```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
%matplotlib widget
```

## Data Pre-Processing
To perform the data processing for each implementation in an easier way, the categorical values are encoded as integer numbers:

| crash | collision | wrongway |
| :-: | :-: | :-: |
| 1 | 2 | 3 |

```python
data = pd.read_csv("simulation_outcomes.csv")
given_solution = data.iloc[0:20,0].values
my_solution = data.iloc[0:20,1].values
```

## Statystical Analysis of Loop Time

```python
# choosing the loop times from the data set
my_solution = np.sort(my_solution)
given_solution = np.sort(given_solution)
my_data1 = my_solution[-7:]
my_data2 = given_solution[-14:]
# plotting histogram for each data set
my_bins1 = [200, 205, 210, 215, 220, 225, 230]
my_bins2 = [170, 175, 180, 185, 190, 195]
fig = plt.figure()
ax = fig.add_subplot(121)
h,e = np.histogram(my_data1, bins=my_bins1)
ax.bar(range(len(my_bins1)-1),h, width=1, edgecolor='k')
ax.set_xticks(range(len(my_bins1)-1))
ax.set_xticklabels(my_bins1[:-1])
ax.title.set_text('my_solution')
ax = fig.add_subplot(122)
h,e = np.histogram(my_data2, bins=my_bins2)
ax.bar(range(len(my_bins2)-1),h, width=1, edgecolor='k')
ax.set_xticks(range(len(my_bins2)-1))
ax.set_xticklabels(my_bins2[:-1])
ax.title.set_text('given_solution')
# finding the statistical values for each data set
avg = statistics.mean(my_data1)
stdev = statistics.stdev(my_data1)
min_value = np.min(my_data1)
max_value = np.max(my_data1)
print("my_solution:\n avg: %5.2f\n stdev: %5.2f\n min: %5.2f\n max: %5.2f\n" % (avg, stdev, min_value, max_value))
avg = statistics.mean(my_data2)
stdev = statistics.stdev(my_data2)
min_value = np.min(my_data2)
max_value = np.max(my_data2)
print("given_solution:\n avg: %5.2f\n stdev: %5.2f\n min: %5.2f\n max: %5.2f" % (avg, stdev, min_value, max_value))
```

## Statystical Analysis of Failure

```python
labels = ["crash", "collision", "wrong way", "success"]
crash = np.count_nonzero(my_solution == 1)
collision = np.count_nonzero(my_solution == 2)
wrong_way = np.count_nonzero(my_solution == 3)
success = np.size(my_solution) - crash - collision - wrong_way
sizes1 = [crash, collision, wrong_way, success]
crash = np.count_nonzero(given_solution == 1)
collision = np.count_nonzero(given_solution == 2)
wrong_way = np.count_nonzero(given_solution == 3)
success = np.size(my_solution) - crash - collision - wrong_way
sizes2 = [crash, collision, wrong_way, success]
fig = plt.figure()
ax = fig.add_subplot(121)
ax.pie(sizes1, labels=labels, autopct='%1.1f%%', startangle=90)
ax.title.set_text('my_solution')
ax = fig.add_subplot(122)
ax.pie(sizes2, labels=labels, autopct='%1.1f%%', startangle=90)
ax.title.set_text('given_solution')
plt.show()
```

## Conclusion
It can be concluded that the given solution's overall performance was better in the following factors:
1. Loop Time: Obviously, the average time required for the robot to finish the loop using the given solution is better than the other one. Also, the standard deviation for the given solution data set much lower, this means that the given solution is more reliable than my solution to finish the loop within the average time, if no failure occures.
2. Failure Percentage: It can be seen from the pie graph, that the given solution has successfully finished the loop in %70 of the iterations but the results for the other solution was the half of this amount. Also, while running the given solution's code, there was no crash occuring but in my solution in %20 of the iterations program crashes.

Considering the collision factor, there was no occurance in my solution but the given solution it happened once so it can be neglected.
